```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Indian_kids_Screen_Time.csv')

# print(df)

df =df.dropna(subset = ['Age', 'Gender', 'Avg_Daily_Screen_Time_hr', 'Primary_Device', 'Exceeded_Recommended_Limit','Educational_to_Recreational_Ratio', 'Health_Impacts', 'Urban_or_Rural'] )


Gender_count = df['Gender'].value_counts()
fig , ax = plt.subplots(1,2, figsize = (12,5))
ax[0].bar(Gender_count.index, Gender_count.values , color = ['skyblue', 'orange'])
ax[0].set_title('Number of Male VS Female')
ax[0].set_xlabel('Gender')
ax[0].set_ylabel('Number of Male and Female')



ax[1].pie(Gender_count,labels = Gender_count.index , autopct = '%1.1f%%' , startangle = 90)
ax[1].set_title('Gender Proportion')

fig.suptitle('Gender analysis')
plt.tight_layout()
plt.savefig('Gender_analysis.png')
plt.show()



plt.figure(figsize = (8,6))
plt.hist(df['Age'], bins = 5, color = 'blue', edgecolor = 'black', label = 'Age group anlysis')
plt.title('Kids Age Group Anlysise')
plt.xlabel('age')
plt.ylabel('Kid age group anlysis')
plt.tight_layout()
plt.savefig('Kids Age Group.png')
plt.show()


Primary_Divices = df['Primary_Device'].value_counts()
plt.figure(figsize = (8,6))
plt.pie(Primary_Divices, labels = Primary_Divices.index , autopct = '%1.1f%%' , startangle = 90)
plt.title('Device Use By Kids')
plt.tight_layout()
plt.savefig('Kids_Devices_use.png')
plt.show()

Health_Impact = df['Health_Impacts'].value_counts()
plt.figure(figsize = (8,6))
plt.pie(Health_Impact, labels = Health_Impact.index , autopct = '%1.1f%%', startangle = 90)
plt.title('Health Issues')
plt.tight_layout()
plt.savefig('Health_Issuse.png')
plt.show()

Urban_and_Rural = df['Urban_or_Rural'].value_counts()
plt.figure(figsize = (8,6))
plt.pie(Urban_and_Rural,labels = Urban_and_Rural.index, autopct = '%1.1f%%', startangle = 90)
plt.title('Kids Belong Area')
plt.tight_layout()
plt.savefig('Kids_Belong_Area.png')
plt.show()



```

```python


Female_data = df[df['Gender'] == 'Female']['Age'].value_counts()
fig, ax = plt.subplots(4,2, figsize=(17,20))                       
ax[0,0].bar(Female_data.index , Female_data.values, color = 'orange')
ax[0,0].set_title('Female Age Group')
ax[0,0].set_xlabel('Age of Female')
ax[0,0].set_ylabel('Number of Female')



Gender_counts = df[df['Gender'] == 'Male']['Age'].value_counts()

ax[0,1].bar(Gender_counts.index, Gender_counts.values, color = 'orange')
ax[0,1].set_title('Male Age Group')
ax[0,1].set_xlabel('Age')
ax[0,1].set_ylabel('Number Of Male')


Age_screen = df[df['Gender'] == 'Female']['Primary_Device'].value_counts()
ax[1,0].bar(Age_screen.index , Age_screen.values, color = 'purple')
ax[1,0].set_title('Device Use')
ax[1,0].set_xlabel('Primary_Device')
ax[1,0].set_ylabel('Number of Female')


Age_screen = df[df['Gender'] == 'Male']['Primary_Device'].value_counts()
ax[1,1].bar(Age_screen.index , Age_screen.values, color = 'purple')
ax[1,1].set_title('Device Use')
ax[1,1].set_xlabel('Primary_Device')
ax[1,1].set_ylabel('Number of Male')

Health_Impacts = df[df['Gender'] == 'Female']['Health_Impacts'].value_counts()
ax[2,0].pie(Health_Impacts, labels = Health_Impacts.index,autopct = '%1.1f%%')
ax[2,0].set_title('Female Health Disease')

Health_Impacts = df[df['Gender'] == 'Male']['Health_Impacts'].value_counts()
ax[2,1].pie(Health_Impacts, labels = Health_Impacts.index,autopct = '%1.1f%%')
ax[2,1].set_title('Male Health Disease')

Urban_or_Rural = df[df['Gender'] == 'Female']['Urban_or_Rural'].value_counts()
ax[3,0].pie(Urban_or_Rural, labels = Urban_or_Rural.index,autopct = '%1.1f%%')
ax[3,0].set_title('Female Belong Area')

Urban_or_Rural = df[df['Gender'] == 'Male']['Urban_or_Rural'].value_counts()
ax[3,1].pie(Urban_or_Rural, labels = Urban_or_Rural.index,autopct = '%1.1f%%')
ax[3,1].set_title('Male Belong Area')



plt.suptitle('Female Data Anlysis')
plt.tight_layout()
plt.savefig('Female_VS_Male_Kids_Analysis.png')
plt.show()
```
